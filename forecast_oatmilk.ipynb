```python
#importae biblioteca e carregar os dados
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df=pd.read_csv('/content/multiTimeline.csv',skiprows=1)
```

```python
print(df.columns)
```

```python
#exemplo de renomeacao
df.rename(columns={df.columns[0]:'data',df.columns[1]:'leite_aveia'},inplace=True)
```

```python
#converter a coluna data para datetime
df['data']=pd.to_datetime(df['data'])
```

```python
#definir a coluna data como indice
df.set_index('data',inplace=True)
```

```python
#teste
df.info()
df.head()
```

```python
#criar colunas de ano e mes para analise sazonal
df['ano']=df.index.year
df['mes']=df.index.month
```

```python
#criar uma tabela mes x ano com a media do interesse
tendencia_sazonal=(df.pivot_table(values='leite_aveia',index='mes',columns='ano'))
```

```python
aggfunce='mean'
```

```python
plt.figure(figsize=(12,6))
tendencia_sazonal.plot(marker='o',colormap='plasma',figsize=(12,6))
plt.title('Sazonalidade do Interesse por Leite de Aveia (média mensal por ano)',fontsize=14)
plt.xlabel('Mes')
plt.ylabel('Interesse médio')
plt.xticks(range(1,13))
plt.grid(True)
plt.tight_layout()
plt.show()
```
